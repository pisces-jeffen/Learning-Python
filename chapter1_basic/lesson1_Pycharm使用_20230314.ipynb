## 上课须知

1. 整理笔记：

   - 笔记一定要自己整理
   - 切记上课的过程中整理笔记
2. 课程：

   - 上课听思路，下课记笔记
   - 上课听思路，下课敲代码
   - 上课过程当中不要剧透
     - 不要用后面的知识点解决前面的问题
     - 没学到的知识点不要提
3. 预习资料：
   - 上个班级的回放
   - 不会统计
4. 课堂知识点：
   - 如果有不理解的地方先扣2->接着扣问题
5. 问题答疑：
   - 问题发在群里(答疑老师或萧何老师，班主任会提示萧何老师)
   - 直接联系萧何（周末或超过5小时没恢复，直接拨打语言电话）
   - 问题比较急，直接打语音
   - 如果问题不急，全部整理在一起，在第二节课堂解答。
   - 上班：周一到周六（有时候周六放假）下午一点半至晚上十点半
6. 课后作业：
   - 必做题：难度相对较小
   - 选做题：涉及思路，难度较高
   - 交作业时间：当天下课结束到第二次课程开始
